## Pythonの制御構文

### for文による繰り返し

九九の3の段と4の段を出力するには、どのようなプログラムを書けばいいでしょうか？

```python
print(3 * 1)
print(3 * 2)
print(3 * 3)
print(3 * 4)
print(3 * 5)
print(3 * 6)
print(3 * 7)
print(3 * 8)
print(3 * 9)

print(4 * 1)
print(4 * 2)
print(4 * 3)
print(4 * 4)
print(4 * 5)
print(4 * 6)
print(4 * 7)
print(4 * 8)
print(4 * 9)
```

他の段を出力するかもしれないので、より汎用的なプログラムに修正しましょう。

```python
x = 3

print(x * 1)
print(x * 2)
print(x * 3)
print(x * 4)
print(x * 5)
print(x * 6)
print(x * 7)
print(x * 8)
print(x * 9)

x = 4

print(x * 1)
print(x * 2)
print(x * 3)
print(x * 4)
print(x * 5)
print(x * 6)
print(x * 7)
print(x * 8)
print(x * 9)
```

ここでprint文が重複していることに注目します。プログラムが長くなると見にくいので、まとめることができると短くなり嬉しいです。このような場合、Pythonのfor文を使うことで「繰り返し」を行うことができます。

```python
for x in [3, 4]:
  print(x * 1)
  print(x * 2)
  print(x * 3)
  print(x * 4)
  print(x * 5)
  print(x * 6)
  print(x * 7)
  print(x * 8)
  print(x * 9)
```

これでリストの要素の3と4に対してprint文を実行することができます。

要素数が増える場合、range()関数を使うことで記述を短くすることができます。

```python
for x in range(3, 5):
  print(x * 1)
  print(x * 2)
  print(x * 3)
  print(x * 4)
  print(x * 5)
  print(x * 6)
  print(x * 7)
  print(x * 8)
  print(x * 9)
```

1の段から9の段まで実行する場合など要素数が増えると、より威力を発揮します。

```python
for x in range(1, 10):
  print(x * 1)
  print(x * 2)
  print(x * 3)
  print(x * 4)
  print(x * 5)
  print(x * 6)
  print(x * 7)
  print(x * 8)
  print(x * 9)
```

先ほどの「[3, 4]」のように「[」と「]」で囲んで記述するデータはリストです。リストはいくつかの変数をまとめたデータ構造です。Pythonではリスト以外に、辞書、タプル、集合といったデータ構造を扱うことができます。データを保持する形式がいくつかあることを知っておきましょう。

```python
l = [1, 2, 3]
print(l)

d = {1: "ito", 2: "ueda", 3: "sato"}
print(d)

t = (1, "ito")
print(t)

s = {1, 2, 3}
print(s)
s = set([1, 2, 3])
print(s)
```

### if文による条件分岐

繰り返しと同じくらいの頻度で人間が行う処理に条件分岐があります。Pythonのif文を使うことで「条件分岐」を行うことができます。例えば九九のテストの点数が60点以上なら合格、それ以外なら不合格とするプログラムを書きます。

```python
score = 86

if score >= 60:
  print("合格")
else:
  print("不合格")
```

```python
score = 59

if score >= 60:
  print("合格")
else:
  print("不合格")
```

ifに続けて記述する条件に一致するとifのブロック、それ以外はelseのブロックを実行します。

条件を続けて記述する場合、elifを使います。

```python
score = 86

if score == 100:
  print("満点")
elif score >= 60:
  print("合格")
else:
  print("不合格")
```

elifを使う場合と、ifを続ける場合の違いは何でしょうか？

```python
score = 86

if score == 100:
  print("満点")
if score >= 60:
  print("合格")
else:
  print("不合格")
```

scoreが86のとき、実行結果は同じです。scoreが100のときはどうでしょうか？

```python
score = 100

if score == 100:
  print("満点")
if score >= 60:
  print("合格")
else:
  print("不合格")
```

「満点」と「合格」が出力されました。1つ目のifを実行した後、2つ目のifも実行されるためです。一方、elifはifに続けて記述する条件に一致しないときのみ条件分岐を行います。

## Pythonの関数とメソッド

### 関数

プログラミングにおける関数は、数学における関数と同じ意味です。すなわち関数は、ある入力に対して何らかの処理を行ってその結果を返します。例えば先ほどのテストの点数に応じて合格か不合格化を条件分岐する、すなわち「採点する（grade）」関数を定義します。かっこの中に入力を受け取るための引数を記述します。

```python
def grade(score):
  if score >= 60:
    return "合格"
  else:
    return "不合格"
```

関数を定義したら、定義した関数名を記述して、必要に応じて引数を記述すると、定義した関数を実行することができます。

```python
score = 86
print(grade(score))
```

関数を使う場合と使わない場合を比較してみましょう。考えるポイントは「修正するときにどちらが修正しやすいか」です。関数を使わないと採点している部分をプログラム全体から探す必要がありますが、関数を使うことで「grade（採点する）」と検索すればすぐに修正することができます。

```python
score = 86

if score >= 60:
  print("合格")
else:
  print("不合格")
```

### メソッド

Pythonには関数と似ているメソッドという処理があります。関数を使うときはprint()やgrade()のように単独で記述するのに対して、メソッドを使うときはscore.append()のように変数に続けて記述します。関数もメソッドも入力を引数に記述するのは同じである一方、メソッドは処理の対象を指定するため変数に続けて記述します。

```python
score = [80, 70, 34, 90, 85]
score.append(40)
print(score)
```

## Pythonのクラス

クラスとインスタンス.png

関数やメソッドが処理をまとめるのに対して、クラスはものの属性をまとめます。またものの属性をまとめたテンプレートであるクラスに対して、クラスから生成される実体はインスタンスです。例えば生徒（Student）というテンプレートから、太郎（taro）という実体を生成するプログラムを書きます。

```python
class Student():
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def take_exam(self, score):
    self.score = score
  
  def announce_score(self):
    print(self.score)
```

定義したStudentクラスからtaroインスタンスを生成して、テストを受け、点数を表示します。

```python
taro = Student("Taro", 18)
taro.take_exam(70)
taro.announce_score()
```

ここで2行目と3行目に注目すると変数に続けて処理が記述されています。これは先ほどのメソッドに該当します。すなわちクラスに紐づく関数がメソッドというわけです。クラスに紐づくということは、クラスから生成されるインスタンスに対してメソッドにより処理を行います。リストの場合と比較してみましょう。

```python
score = [80, 70, 34, 90, 85]
score.append(40)
print(score)
```

taroインスタンス以外にも、複数のインスタンスを生成することが可能です。jiroインスタンスとsaburoインスタンスを生成してみましょう。

```python
jiro = Student("Jiro", 17)

saburo = Student("Saburo", 15)
```
