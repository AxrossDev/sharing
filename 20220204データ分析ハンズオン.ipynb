### データの前処理

オープンデータ含むデータにはしばしばノイズが含まれています。
ここでいうノイズとはデータの「欠け」や「偏り」のことを指します。
データの分析に入る前に、この「欠け」や「偏り」を補う処理が必要です。
この処理を前処理と呼びます。

前処理では、「欠け」となっている欠損値や「偏り」となっている外れ値を検出して補います。
今回のデータである「新規陽性者数」と「重症患者数」について欠損値や外れ値の検出を行いましょう。

まずは「新規陽性者数」のデータに対して前処理を行います。
欠損値を確認するには次のコードを実行します。

```python
df_patients.isnull().sum()
```

各列にいくつnull、つまり空の値、欠損値が入っているかがわかります。
欠損値が0の列はすべて値が埋まっている理想の状態です。
出力を見ると No, 全国地方公共団体コード, 都道府県名, 公表_年月日, 患者_年代, 患者_性別 の列には欠損値がないことがわかります。
今回は新規陽性者数の日付ごとの推移を確認したいのですが、欠損値がない「公表_年月日」を日付の値とすれば欠損値を補うことなく分析することができます。
そのため「公表_年月日」を使用しましょう。
もし欠損値を含む列を使用する場合は補完する必要がありますが、本レシピでは割愛します。
気になる方は「pandas 欠損値 補完」と調べてみてください。

「欠け」である欠損値の処理を終えたら、次は「偏り」である外れ値の処理を行います。
何をもって外れ値とするかは対象データによります。
例えば人の身長のデータをcm単位で表している場合、「1.6」というデータがあったらmとcmを誤っていて本来は「160」と想定され、「1800」というデータがあったらmmとcmを誤っていて本来は「180」と想定されますね。
このように人の身長で「1.6」や「1800」というデータがあれば外れ値だろうと見当がつきます。
またお菓子の価格のデータを円単位で表している場合、「3」というデータがあったら$と円を誤っていると想定され外れ値だろうと見当がつきます。
このように対象データにより外れ値かどうか判断するため、対象データの性質、特に小さくてもこの値、大きくてもこの値、という感覚を持っておきましょう。

それでは外れ値の検出を行います。
まず今回は日付ごとの新規陽性者数をカウントするため、日付による集計を行います。
「日ごと」の新規陽性者数を「カウント」する場合、groupbyメソッドを用いて行うことができます。
groupbyの引数には「日ごと」＝公表_年月日を与え、groupbyメソッドに続けて「カウント」＝countと記述します。
count以外だと数値を合計するときにはsum、平均をとるときにはmeanと記述します。

```python
df_patients = df_patients.groupby(['公表_年月日']).count()
df_patients
```

同じ日付が入っている行の合計数を求めることができました。
元のデータは1行あたり1名の新規陽性者数に関する情報が入っていたため、このコードで日付あたりの新規陽性者数を求められるというわけです。
同じ値が入っている列が複数あっても仕方がないので、先ほど欠損値の確認をしたときに欠損値がなかった列の1つ「No」を使用することにします。

```python
df_patients = df_patients.reset_index()[['公表_年月日', 'No']]
df_patients
```

これで日付あたりの新規陽性者数をきれいに集計できましたね。

ここで外れ値の確認です。
外れ値を確認するには次のコードを実行します。

```python
df_patients.describe()
```

describeメソッドを使用すると行数、平均値、標準偏差、最小値、四分位数、最大値を算出できます。
最小値が1となっていて正の値になっていること、また最大値が17631と最近ニュースで聞いた値になっていることから、今回のデータには外れ値がないことがわかります。

### 複数データの可視化と関係の分析

今回のテーマは「新規陽性者数と重症患者数の関係」を可視化により分析することでした。
このような複数データの可視化について解説します。

考え方として出力したいグラフから逆算していきます。
今回はx軸に日付を置き、新規陽性者数と重症患者数を2つの折れ線グラフで可視化することを想定します。
なぜ折れ線グラフかというと、棒グラフだと重なってしまう部分が見にくい、また変化の傾向が見にくいためです。

このグラフを出力するために「日付」「新規陽性者数」「重症患者数」の3列から成るDataFrameを作成します。
そのためには新規陽性者数のDataFrameであるdf_patientsと、重症患者数のDataFrameであるdf_seriousを同じ日付で結合します。

先にそれぞれのDataFrameの行数を確認します。

```python
print(df_patients.shape)
print(df_serious.shape)
```

出力からdf_patientsは714行、df_seriousは705行あることがわかります。
ここで行数が違うということは片方には含まれているがもう片方にはない日付が存在するということです。
このことに気をつけながら結合を行います。
DataFrameの結合にはpd.merge関数を使用します。

```python
df_merge = pd.merge(df_patients, df_serious, how='inner')
df_merge
```

ここで引数how='inner'と指定することで、両方のDataFrameに含まれる「日付」の値のみ残すことができます。
他には第一引数に指定するDataFrameに合わせる'left', 第二引数に指定するDataFrameに合わせる'right', どちらかに含まれる値を残す'outer'を指定できます。
ここで結合元のDataFrameに値がない場合はNaNが入ります。
気になる方はhowに与える値を変えてみてください。

今回は結合の結果、700行のDataFrameになりました。
これを折れ線グラフで可視化します。
先ほどと同様にPlotlyを使用しますが、複数グラフのプロットのため使用する関数が異なります。

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_merge['公表_年月日'], y=df_merge['No'], name='新規陽性者数'))
fig.add_trace(go.Scatter(x=df_merge['公表_年月日'], y=df_merge['重症'], name='重症患者数'))
fig.show()
```

イメージとして、グラフを1つずつ足していくように複数グラフをプロットします。
まずgo.Figure()を定義して、そこにadd_traceメソッドでグラフを追加していきます。
add_traceメソッドの引数にはgo.Scatterを与え、そのgo.Scatterの引数にはx軸の値、y軸の値、プロットの名前、の順に指定します。

さて、新規陽性者数と重症患者数をそれぞれプロットしたものの、重症患者数の推移がよくわかりません。
これは最小値が1で最大値が17000を超える新規陽性者数と、最小値が1で最大値が300もいかない重症患者数を同じy軸でプロットしてしまっているためです。
重症患者数のプロットにもう1つy軸を追加しましょう。

```python
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{'secondary_y': True}]])

fig.add_trace(go.Scatter(x=df_merge['公表_年月日'], y=df_merge['No'], name='新規陽性者数'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_merge['公表_年月日'], y=df_merge['重症'], name='重症患者数'), secondary_y=True)
fig.update_yaxes(range=[0, 20000], secondary_y=False)
fig.update_yaxes(range=[0, 500], secondary_y=True)
fig.show()
```

plotly.subplots.make_subplots関数を用いてsecondary_yによりどちらのy軸に沿ってプロットするかを選択しています。
今回は新規陽性者数を1つ目のy軸に0～20,000の範囲で、重症患者数を2つ目のy軸に0～500の範囲でプロットしています。
出力を見ると先ほどより重症患者数の推移がはっきりしましたね。

### データの前処理

### 地図上への可視化

それでは区市町村ごとの陽性者数を緯度・経度との組み合わせにより地図上へ可視化します。
考え方として出力したいグラフから逆算します。

次は緯度・経度で指定するために先ほど絞ったDataFrameと緯度・経度のDataFrameを組み合わせます。

ここで引数how='inner'と指定することで、両方のDataFrameに含まれる「市区町村名」の値のみ残すことができます。
他には第一引数に指定するDataFrameに合わせる'left', 第二引数に指定するDataFrameに合わせる'right', どちらかに含まれる値を残す'outer'を指定できます。

```python
import plotly.express as px

fig = px.scatter_mapbox(df, lat='緯度', lon='経度')
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

地図が表示され、いくつか点がプロットされました。

px.scatter_mapboxの引数にsize='陽性者数'を加えました。
すると先ほどは各点が同じサイズの点だったのに対して、今回は点のサイズが小さいものから大きいものまで出てきました。
そしてなんとなく23区内は点が大きく、その他は小さくなっています。
このサイズは陽性者数によって決まっています。
陽性者数の値が大きければ点も大きくなり、陽性者数の値が小さければ点も小さくなるわけです。
これなら陽性者数の違いがわかりやすいですね。

color='陽性者数'を追加しました。

「東京都_新型コロナウイルス陽性患者発表詳細」のデータを再度読み込みます。

```python
import pandas as pd

df = pd.read_csv('130001_tokyo_covid19_patients.csv')
df
```

### データの前処理

今回のデータについても欠損値や外れ値の検出を行います。

まずは欠損値の確認から行います。

```python
df.isnull().sum()
```

今回は年代別の新規陽性者数の推移を可視化したいため、「患者_年代」に注目します。
「患者_年代」の欠損値の数は0、つまり全ての値がきちんと入っています。

次に外れ値を確認します。
データを見ると「患者_年代」は「X0代」と表記されるようです。
しかし思いもよらない誤記があるかもしれないため、value_countsメソッドを使用して値ごとの要素数を数えます。

```python
df['患者_年代'].value_counts()
```

年代の表記として一部「-」「不明」のものがありますが、それ以外は妥当な値です。
「-」と「不明」を削除してしまうと年代がわからない方もいるという事実を捻じ曲げてしまうため、そのまま残しておきます。

### 年代別の新規陽性者数の推移

それではいよいよ分析を行いましょう。

もう一度、各年代の要素数を数えてみます。

```python
df['患者_年代'].value_counts()
```

全期間で各年代の新規陽性者数がどれくらいいるか集計できました。

今回はここからさらにもう一歩、年代別の新規陽性者数の「推移」を分析します。
日ごとの推移を分析するにはgroupbyメソッドを用いて行い、groupbyの引数には「日ごと」＝公表_年月日、「各年代ごと」＝患者_年代を与え、groupbyメソッドに続けて「カウント」＝countと記述します。

```python
df_groupby = df.groupby(['公表_年月日', '患者_年代']).count()
df_groupby
```

左にインデックスとして「公表_年月日」と「患者_年代」が、表の値としてデータ数のカウントが集約されました。
これで年代別の新規陽性者数の推移が見れそうなので見やすくするために可視化を行います。
可視化のため、インデックスとなっている「公表_年月日」と「患者_年代」を列に直します。
reset_indexメソッドを使うとインデックスをリセット、つまり0から始まるユニークな番号で表すことができます。
さらにプロットのときに必要な値のみに絞っておきます。

```python
df_groupby = df_groupby.reset_index()[['公表_年月日', '患者_年代', 'No']]
df_groupby
```

これで準備ができたので可視化を行いましょう。
今回は推移を見たいため、折れ線グラフを描画します。
px.line関数の引数xに「公表_年月日」、yに「No」を指定すると各日付あたりの新規陽性者数を表示できます。
そしてcolorに「患者_年代」を指定すると各年代ごとにグラフを色分けして表示できます。

```python
import plotly.express as px

fig = px.line(df_groupby, x='公表_年月日', y='No', color='患者_年代')
fig.show()
```

```python
import plotly.express as px

fig = px.line(df_groupby, x='公表_年月日', y='No', color='患者_年代',
              category_orders={'患者_年代': ['100歳以上', '90代', '80代', '70代', '60代', '50代', '40代', '30代', '20代', '10代', '10歳未満', '-', '不明']})
fig.show()
```

colorに指定した「患者_年代」は右に一覧で表示されます。
各年代をクリックするとその年代を表示、もしくは非表示にできます。
これにより特定の年代を必要なだけ選んで比較することができます。

例えば20代と70代のみ選択すると、どちらも増加の山が2020年4月、2020年8月、2021年1月あたりにあり、一方で2021年8月については20代は増加の山があるが70代は落ち着いていることがグラフから読み取れます。

これはワクチンが普及したことによると考えられます。

しかし2022年1月はどちらも山があり、ワクチンの効果が徐々に弱くなってきているか、帰省により20代と70代が一緒に過ごすことが多かったからか、などいろんな原因が考えられます。

また10歳未満と20代のみ選択すると、10歳未満は2020年4月、2020年8月に増加の山がない、一方で2021年1月、2021年8月、2022年1月には増加の山があることがグラフから読み取れます。

いわゆる変異株の影響によるもので、変異株は10歳未満へも感染、発症するものと考えられます。

他にも読み取れることはたくさんあるので、ぜひ年代を自由に選択して比較してみてください。
