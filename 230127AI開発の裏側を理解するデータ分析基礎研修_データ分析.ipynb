```python
import pandas as pd

df = pd.read_csv('ecommerce.csv')
df
```

```python
pd.set_option('display.max_columns', 50)
df
```

```python
df['Product Category'].unique()
```

```python
df = df[df['Product Category'] == 'Fashion']
df
```

```python
df['Product'].unique()
```

```python
df = df[['Product', 'Quantity', 'Sales']]
df
```

```python
df.info()
```

```python
df['Quantity'].unique()
```

```python
df['Quantity'] = df['Quantity'].astype('int')
df['Quantity'].unique()
```

```python
df['Sales'] = df['Sales'].str.replace('$', '')
df['Sales'] = df['Sales'].astype('float')
df = df.rename(columns={'Sales': 'Price'})
df['Price'].unique()
```

```python
df
```

```python
df.to_csv('fashion.csv', index=False)
```

```python
df = df.groupby('Product')['Price','Quantity'].agg({
    'Price':['min','max',],
    'Quantity':'sum'
}).reset_index()
df
```

```python
df = pd.concat([df['Product'], df['Price']['max'], df['Quantity']['sum']], axis=1)
df = df.rename(columns={'max': 'Price', 'sum': 'Quantity'})
df
```

```python
df['Cost'] = df['Price'] * 0.35
df['Quantity'] = df['Quantity'] - 4000
df
```

```python
import numpy as np
import plotly.express as px

x = np.arange(0, 6000)
sales = x * df['Price'][0]
```

```python
fig = px.line(x=x, y=sales)
fig.show()
```

```python
fig = px.line(x=x, y=sales, labels={'x': 'Quantity', 'y': 'Sales'})
fig.show()
```

```python
x = np.arange(0, 6000)
cost = x * df['Cost'][0] + 300000

fig = px.line(x=x, y=cost, labels={'x': 'Quantity', 'y': 'Cost'})
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=sales, name='売上'))
fig.add_trace(go.Scatter(x=x, y=cost, name='コスト'))
fig.update_layout(xaxis_title='Quantity', yaxis_title='Sales/Cost')
fig.show()
```
